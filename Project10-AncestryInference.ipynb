```python
import os
import subprocess

# Define input and output directories
input_dir = "/storage/ice-shared/biol6150/Data/1000Genomes/"
output_dir = "/home/hice1/imusayeva3/biol6150/ProjectSubmissions/Group6/Project10/Plink_Binary"
os.makedirs(output_dir, exist_ok=True)

# Generate VCF file names for chromosomes 1 through 22
vcf_files = [f"phase3.chr{i}.GRCh38.GT.crossmap.vcf.gz" for i in range(1, 23)]

# Loop through each VCF file and convert it directly to binary PLINK format
for vcf_file in vcf_files:
    vcf_path = os.path.join(input_dir, vcf_file)
    binary_prefix = os.path.join(output_dir, vcf_file.replace(".vcf.gz", "_binary"))
    
    # Command to run PLINK for direct conversion
    plink_command = [
        "plink",
        "--vcf", vcf_path,      
        "--make-bed",        
        "--out", binary_prefix  
    ]
    
    print(f"Converting {vcf_file} to binary PLINK format...")
    subprocess.run(plink_command, check=True)
    print(f"Converted {vcf_file} -> {binary_prefix}")

print("All VCF files have been directly converted to binary PLINK format.")
```

```python
import os
import subprocess

# Define directories
input_dir = "/home/hice1/imusayeva3/biol6150/ProjectSubmissions/Group6/Project10/Plink_Binary"
qc_output_base_dir = "/home/hice1/imusayeva3/biol6150/ProjectSubmissions/Group6/Project10/QC_Output"
os.makedirs(qc_output_base_dir, exist_ok=True)

# Generate binary PLINK file prefixes for chromosomes 1 through 22
binary_files = [f"phase3.chr{i}.GRCh38.GT.crossmap_binary" for i in range(1, 23)]

# Loop through each PLINK binary file and perform QC
for binary_file in binary_files:
    input_prefix = os.path.join(input_dir, binary_file)
    output_dir = os.path.join(qc_output_base_dir, binary_file) 
    os.makedirs(output_dir, exist_ok=True)  
    output_prefix = os.path.join(output_dir, binary_file + "_QC")
    
    # Command to run PLINK for QC
    plink_qc_command = [
        "plink",
        "--bfile", input_prefix,  
        "--mind", "0.05",         
        "--geno", "0.05",         
        "--maf", "0.01",          
        "--hwe", "1e-6",          
        "--make-bed",             
        "--out", output_prefix    
    ]
    
    print(f"Running QC on {binary_file}...")
    subprocess.run(plink_qc_command, check=True)
    print(f"QC completed for {binary_file}. Output saved to {output_dir}")

print("QC steps completed for all PLINK binary files.")
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import os

# Define the directory containing the .Q files
data_dir = "/home/hice1/imusayeva3/biol6150/ProjectSubmissions/Group6/Project10/Admixture_After_QC"

# Define the chosen K values for each chromosome
chosen_k_values = {
    "chr1": 10, "chr2": 5, "chr3": 8, "chr4": 7, "chr5": 9,
    "chr6": 10, "chr7": 9, "chr8": 10, "chr9": 8, "chr10": 10,
    "chr11": 10, "chr12": 10, "chr13": 10, "chr14": 10, "chr15": 10,
    "chr16": 10, "chr17": 10, "chr18": 10, "chr19": 10, "chr20": 10,
    "chr21": 9, "chr22": 10
}

top_individuals_data = {}
for chrom, k in chosen_k_values.items():
    q_file = f"{data_dir}/phase3.{chrom}.GRCh38.GT.crossmap_binary_QC.{k}.Q"    
    if os.path.exists(q_file):
        df = pd.read_csv(q_file, sep=" ", header=None)        
        df["Individual"] = [f"Ind_{i+1}" for i in range(df.shape[0])]        
        top_10 = df.iloc[:, :-1].max(axis=1).sort_values(ascending=False).head(10).index  
        top_individuals_data[chrom] = df.iloc[top_10]  
    else:
        print(f"File not found: {q_file}")

# Plot the results for each chromosome
for chrom, top_data in top_individuals_data.items():
    top_data = top_data.set_index("Individual")    
    plt.figure(figsize=(10, 6))
    top_data.iloc[:, :-1].plot(kind="bar", stacked=True, colormap="tab10", figsize=(12, 6))
    plt.title(f"Top 10 Individuals' Ancestry Proportions - {chrom} (K={chosen_k_values[chrom]})", fontsize=16)
    plt.xlabel("Individuals", fontsize=14)
    plt.ylabel("Ancestry Proportion", fontsize=14)
    plt.xticks(rotation=45, ha="right")
    plt.legend(title="Ancestry Components", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

```

```python
import os
import pandas as pd

# Define directories and file paths
data_dir = "/home/hice1/imusayeva3/biol6150/ProjectSubmissions/Group6/Project10/Admixture_After_QC"
metadata_file = "/home/hice1/imusayeva3/biol6150/ProjectSubmissions/Group6/Project10/kgp3_id_pop.tsv"

# Chosen K values for each chromosome
chosen_k_values = {
    "chr1": 10, "chr2": 5, "chr3": 8, "chr4": 7, "chr5": 9,
    "chr6": 10, "chr7": 9, "chr8": 10, "chr9": 8, "chr10": 10,
    "chr11": 10, "chr12": 10, "chr13": 10, "chr14": 10, "chr15": 10,
    "chr16": 10, "chr17": 10, "chr18": 10, "chr19": 10, "chr20": 10,
    "chr21": 9, "chr22": 10
}

# Load metadata
if not os.path.exists(metadata_file):
    raise FileNotFoundError(f"Metadata file not found: {metadata_file}")

metadata = pd.read_csv(metadata_file, sep="\t", names=["Individual", "Population"])

# Initialize a dictionary to store average ancestry results
average_ancestries = {}

# Process each chromosome
for chrom, k in chosen_k_values.items():
    q_file = f"{data_dir}/phase3.{chrom}.GRCh38.GT.crossmap_binary_QC.{k}.Q"    
    if not os.path.exists(q_file):
        print(f"File not found: {q_file}")
        continue    
    df = pd.read_csv(q_file, sep=" ", header=None)
    df["Individual"] = metadata["Individual"].iloc[:df.shape[0]]      
    merged_data = pd.merge(df, metadata, on="Individual", how="inner")    
    ancestry_data = merged_data.drop(columns=["Individual", "Population"])    
    ancestry_data["Population"] = merged_data["Population"]    
    grouped = ancestry_data.groupby("Population").mean()    
    average_ancestries[chrom] = grouped

# Combine results across chromosomes
final_averages = pd.concat(average_ancestries.values(), keys=average_ancestries.keys(), axis=1)

# Save results to a CSV file
output_file = f"{data_dir}/average_ancestry_by_population_fixed.csv"
final_averages.to_csv(output_file)
print(f"Average ancestry values saved to: {output_file}")
```
